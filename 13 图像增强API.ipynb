```python
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow

name = './gugong.jpg'
img_string = tf.read_file(name) #以字符串的方式读取图片
img_decoded = tf.image.decode_image(img_string) #解析图片
sess = tf.Session()
img_decoded_val = sess.run(img_decoded)
print(img_decoded_val.shape)
%matplotlib inline
imshow(img_decoded_val)
```

```python
# resize

# tf.image.resize_area
# tf.image.resize_bicubic #二次线性函数方法
# tf.image.resize_nearest_neighbor #用最邻近像素放大

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 365, 600, 3])##这个函数适应四维因为是mini_batch

resize_img = tf.image.resize_bicubic(img_decoded, [730, 1200])

sess = tf.Session()
img_decoded_val = sess.run(resize_img)
img_decoded_val = img_decoded_val.reshape((730, 1200, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)#计算完以后为float重新变回int值
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# crop
# tf.image.pad_to_bounding_box #把图像放在一个画布上
# tf.image.crop_to_bounding_box #裁剪再放在一个画布上
# tf.random_crop #在原始图像随机裁剪

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 365, 600, 3])

padded_img = tf.image.pad_to_bounding_box( #这个函数适应四维因为是mini_batch
    img_decoded, 50, 100, 500, 800) #50，100原始图像在新画布的位置，500,800画布大小

sess = tf.Session()
img_decoded_val = sess.run(padded_img)
img_decoded_val = img_decoded_val.reshape((500, 800, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# flip
# tf.image.flip_up_down 上下翻转
# tf.image.flip_left_right 左右翻转
# tf.image.random_flip_up_down 随机上下翻转
# tf.image.random_flip_left_right 随机左右翻转

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 365, 600, 3])

flipped_img = tf.image.flip_up_down(img_decoded)

sess = tf.Session()
img_decoded_val = sess.run(flipped_img)
img_decoded_val = img_decoded_val.reshape((365, 600, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# brightness
# tf.image.adjust_brightness 有一个参数，手动设定
# tf.image.random_brightness 随机设定
# tf.image.adjust_constrast
# tf.image.random_constrast

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 365, 600, 3])

new_img = tf.image.adjust_brightness(img_decoded, 0.5)#0.5控制幅度大小 +50%

sess = tf.Session()
img_decoded_val = sess.run(new_img)
img_decoded_val = img_decoded_val.reshape((365, 600, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```
